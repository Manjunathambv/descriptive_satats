```python
# 1
## Start code here
data = pd.read_csv('report.csv')
data
## End code here
```

```python
# 3
## Start code here
# Q3. Find the eldest student. (1 marks)
data[data["Date_of_birth"] == data["Date_of_birth"].min()]

## End code here
```

```python
# 4
## Start code here
import numpy as np
# Q4. Use agg.() method to find (0.5) quantile on Maths column. (1 marks)
data.agg({"Maths":"median"})


## End code here
```

```python
# 5
## Start code here

# Q5. Use agg.() method on multiple columns(Maths & Physics) to find (0.5) quantile. (1/2 marks)
data.agg({"Maths":"median", "Physics":"median"})

## End code here
```

```python
# 6
## Start code here[
# Q6. Use agg.() method on Maths column to find multiple Quantile (0.5 & 0.3). (1/2 marks)
def q1(x):
    return x.quantile(0.5)
def q2(x):
    return x.quantile(0.3)
data.agg({"Maths":[q1, q2]})

## End code here
```

```python
# 7
## Start code here

# Q7. Find cumulative sum of Maths. (1 marks)
data['Maths'].agg({"Maths":"sum"})

## End code here
```

## Grouped Summary Statistics
**(4 marks each)**

Q1. Find mean value by grouping city and maths column.

Q2. Find median value by grouping city , maths & physics column.

Q3. Using .agg method find the min, max & sum value by grouping city, maths & physics column.

#### Many groups, many summaries
Q4. Find mean value by grouping multiple variable (City , Place) & (Maths , Physics)  

```python
# 1
## Start code here
# Q1. Find mean value by grouping city and maths column.
# data.groupby(["City"]).agg({"Maths":"mean"})
temp = data.groupby(["City"])
temp[["Maths"]].mean().add_prefix("mean_score_").reset_index()
## End code here
```

```python
# 2
## Start code here

# Q2. Find median value by grouping city , maths & physics column.
# data.groupby(["City"]).agg({"Maths":'median',
#                            "Physics":'median'}).reset_index()
temp = data.groupby(["City"])
temp[["Maths", "Physics"]].median().add_prefix("median_value_").reset_index()

## End code here
```

```python
# 3
## Start code here
# Q3. Using .agg method find the min, max & sum value by grouping city, maths & physics column.

data.groupby(["City"]).agg({"Maths":['min', 'max', 'sum'],
                           "Physics":['min', 'max', 'sum']}).reset_index()

## End code here
```

```python
# 4
## Start code here
# Q4. Find mean value by grouping multiple variable (City , Place) & (Maths , Physics)
# data.groupby(["City", "Place"]).agg({"Maths":'mean',
#                                     'Physics':'mean'})

temp = data.groupby(["City", 'Place'])
temp[["Maths", "Physics"]].mean().add_prefix("mean_")
## End code here
```

## Pivot Tables
**(4 marks each)**

Q1. Pivot the Report Table by keeping index as "Place" & value as "Chemistry".

Q2. Using **numpy .agg** method find the median of the pivot table in question number 1.

Q3. Using **numpy .agg** method perform multiple statistics to find mean & median of the pivot table in question number 1.

```python
# 1
## Start code here
# data1 = data.set_index(["Place"])
# data1
data1 = pd.pivot_table(data, values="Chemistry", index="Place")
data1
## End code here
```

```python
# 2
## Start code here
# Q2. Using numpy .agg method find the median of the pivot table in question number 1.
data1.agg({'Chemistry':np.median})
## End code here
```

```python
# 3
## Start code here
# Q3. Using numpy .agg method perform multiple statistics to find mean & median of the pivot table in question number 1.
data1.agg({"Chemistry":[np.mean, np.median]})
## End code here
```
